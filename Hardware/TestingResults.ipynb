```python
import numpy as np
import numpy.linalg as la
import numpy.random as npr
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as stats
%matplotlib inline
plt.style.use('bmh')
```

```python
defaultTime = [0,.57, .52, .52, .54, .63, .75, .67, .68, .51, 0]
Multiprocessing = [0,7.77, 7.07, 7.77, 7.25, 7.2, 7.81, 7.25, 7.6, 7.32, 0]
```

```python
plt.plot(range(11), defaultTime, label = 'default time', color = 'black')
plt.title("Local Test (I7-6820HQ @ 2.70GHz)")
plt.xlabel("Test Iterations")
plt.ylabel("Execution Time (s)")
ax = plt.gca()
ax.set_ylim([0,1])
ax.fill(range(11), defaultTime, "black", alpha=.15)
plt.legend()
```

```python
plt.plot(range(11), defaultTime, label = 'default time', color = 'black')
plt.plot(range(11), Multiprocessing, label = 'Multiprocessing', color = 'red')
plt.title("Local Test (I7-6820HQ @ 2.70GHz)")
plt.xlabel("Test Iterations")
plt.ylabel("Execution Time (s)")
ax = plt.gca()
ax.set_ylim([0,8])
ax.fill(range(11), defaultTime, "black", alpha=.15)
ax.fill(range(11), Multiprocessing, "red", alpha=.3)
plt.legend()
```
